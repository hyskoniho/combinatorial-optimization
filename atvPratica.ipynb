```python
import json, random
```

# Construção do Mercado Pão de Açúcar

```python
market_itens = {}
with open(r'.\\data\\market.json', 'r', encoding='utf-8') as file:
    market_itens = json.load(file)

market_itens = {int(k): v for k, v in market_itens.items()}
market_itens
```

```python
carrinho_compras = {
    'Altura': 65,
    'Largura': 50,
    'Comprimento': 80,
    'PesoMaximo': 460*1000
}

carrinho_compras['Capacidade'] = carrinho_compras['Altura'] * carrinho_compras['Largura'] * carrinho_compras['Comprimento']

carrinho_compras
```

# Constraints

```python
valor_maximo = 2000
itens_maximo = 20
```

# Funções

```python
def calcular_volume(item):
    return float(item['Altura']) * float(item['Largura']) * float(item['Profundidade'])
```

# Genético

```python
# Escolher um produto
def gerar_produto(itens_proibidos = []) -> str:
    id_ = random.choice(list(market_itens.keys()))
    return id_ if id_ not in itens_proibidos else gerar_produto(itens_proibidos)
```

```python
# Gera um individuo
def gerar_carrinho() -> list[int]:
    volume_compra = 0 
    valor_compra = 0
    compra = []

    while len(compra) < itens_maximo:
        produto_selecionado = gerar_produto()
        
        volume_compra += calcular_volume(market_itens[produto_selecionado])
        valor_compra += float(market_itens[produto_selecionado]['Preço'])
        compra.append(produto_selecionado)

    return compra
```

```python
compras = [gerar_carrinho() for _ in range(5)]
```

```python
for idx, compra in enumerate(compras):
    print(f"Compra {idx+1}:")
    valor_compra = 0
    volume_compra = 0
    
    for item in compra:
        print(f"Item: {market_itens[item]['Nome']} - R$ {market_itens[item]['Preço']}")
        valor_compra += float(market_itens[item]['Preço'])
        volume_compra += calcular_volume(market_itens[item])
    
    print('\n')
    print(f"Valor total da compra: R${valor_compra}")
    print(f"Volume total da compra: {volume_compra} cm³")
    print('\n')
```
